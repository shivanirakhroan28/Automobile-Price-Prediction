```python
def outlier(x):
    q1= x.quantile(0.25)
    q2= x.quantile(0.50)
    q3= x.quantile(0.75)
    iqr= q3-q1
    upper= q3+(1.5*iqr)
    lower= q1-(1.5*iqr)
    
    return upper,lower
```
